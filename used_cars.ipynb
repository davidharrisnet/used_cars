```python
bins = [0, 500, 1000, 2500, 5000, 10000, 15000, 20000, 25000, 30000,40000, 100000]
```

```python
labels = ['500', '1000', '2500', '5000', '10000', '15000', '20000', '25000', '30000', '40000', '100000']
```

```python
# Create new category column
autodf['PriceRange'] = pd.cut(autodf['price'], bins=bins, labels=labels)

# Example: Group by the new category
summary = autodf.groupby('PriceRange', observed=False)['price'].count()
#print(autodf)
print(summary)
```
